```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def soft(data, n=7):
    start = n // 2
    stop = data.size - start
    parity = n % 2
    medium = [np.mean(data[i - start: i + start + parity]) for i in range(start, stop)]
    begin = [np.mean(data[:2 * i + 1]) for i in range(start)]
    end = [np.mean(data[-2 * i - 1:]) for i in range(start)][::-1]
    return np.array(begin + medium + end)
```

```python
data = np.loadtxt(open('./noise data.csv'), delimiter=',')
```

```python
plt.plot(data, label='initial')
plt.plot(soft(data, n=31), label='softed by 31')
plt.grid()
plt.legend()
plt.show()
```
